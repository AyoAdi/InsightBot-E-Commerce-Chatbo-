### Handling "plet"

```python
plet
```

```python
plet.info()
```

```python
plet['eligibility_datetime_utc'] = pd.to_datetime(plet['eligibility_datetime_utc']) # Converting the column to datetime format from object format
```

```python
plet.info()
```

```python
duplicates_in_item_id = plet['item_id'].duplicated(keep=False)
num_duplicates_item_id = duplicates_in_item_id.sum()
print(num_duplicates_item_id)
```

```python
duplicates_in_composite_key = plet.duplicated(subset=['item_id', 'eligibility_datetime_utc'], keep=False) # Since item_id has duplicates, it means it has to be a composite key with another column
num_duplicates_composite_key = duplicates_in_composite_key.sum()
print(num_duplicates_composite_key) # (item_id, eligibility_datetime_utc) is the composite key
```
